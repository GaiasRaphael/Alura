```python
colunas = list(hospedagem.columns)
colunas
```

```python
hospedagem = hospedagem.explode (colunas[3:]).reset_index(drop=True)
hospedagem
```

```python
hospedagem.reset_index(inplace=True, drop=True)
```

```python
hospedagem
```

```python
import numpy as np
```

```python
hospedagem['max_hospedes'] = hospedagem['max_hospedes'].astype (np.int64)
hospedagem.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
hospedagem[col_numericas] = hospedagem[col_numericas].astype(np.int64)
hospedagem.info()
```

```python
hospedagem['avaliacao_geral'] = hospedagem['avaliacao_geral'].astype(np.float64)
```

```python
hospedagem['preco'] = hospedagem['preco'].apply(lambda x: x.replace('$','').replace(',','').strip()).astype(np.float64)
hospedagem.info()
```

```python
hospedagem['preco']
```

```python
hospedagem[['taxa_deposito','taxa_limpeza']] = hospedagem[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip()).astype(np.float64)
hospedagem.info()
```

TOKENIZAÇÃO

```python
hospedagem['descricao_local'] = hospedagem['descricao_local'].str.lower()
```

```python
hospedagem['descricao_local'] = hospedagem['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
hospedagem['descricao_local'] = hospedagem['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
hospedagem['descricao_local'] = hospedagem['descricao_local'].str.split()
```

```python
hospedagem['descricao_local']
```

```python
hospedagem['comodidades'] = hospedagem['comodidades'].str.replace ('\{|}|\"', '', regex=True)
```

```python
hospedagem['comodidades'] = hospedagem['comodidades'].str.split(',')
hospedagem['comodidades']
```

```python
hospedagem['descricao_vizinhanca'][39]
```

```python
#hospedagem['descricao_vizinhanca'] = hospedagem['descricao_vizinhanca'].str.lower()
hospedagem['descricao_vizinhanca'] = (hospedagem['descricao_vizinhanca']
                                      .str.replace ('[^a-zA-Z0-9\-\']', ' ', regex=True)
                                      .str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
                                      .str.split())
hospedagem['descricao_vizinhanca']
```

```python
disponivel_por_mes = moveis_disponiveis.groupby(moveis_disponiveis['data'])[['vaga_disponivel']].sum()
```

```python
disponivel_por_mes
```
