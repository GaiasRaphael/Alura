```python
colunas_list = list(seeg.loc[:,'Nível 1 - Setor': 'Produto'].columns)
```

```python
emissao_anos = list(seeg.loc[:, 1970:2021].columns)
```

```python
seeg.melt (id_vars= colunas_list, value_vars=emissao_anos, var_name= 'Ano', value_name='Emissão')
```

```python
emissoes_por_ano = seeg.melt (id_vars= colunas_list, value_vars=emissao_anos, var_name= 'Ano', value_name='Emissão')
```

```python
emissoes_por_ano.groupby('Gás').sum(numeric_only=True)
```

```python
emissao_por_gas = emissoes_por_ano.groupby('Gás').sum(numeric_only=True).sort_values ('Emissão', ascending=False)
```

```python
emissao_por_gas
```

```python
emissao_por_gas.plot.barh()
```

```python
emissao_por_gas.iloc[0:9]
```

```python
print (f'A emissão de CO2 corresponde a {float((emissao_por_gas.iloc[0:9].sum() / emissao_por_gas.sum()).iloc[0]) * 100:.2f} % de emissão total')
```
