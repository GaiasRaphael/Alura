```python
populacao = pd.read_excel ('/content/POP2022_Municipios.xls', header=1, skipfooter=34)
populacao.tail()
```

```python
populacao.groupby ('UF').sum (numeric_only = True)
```

```python
populacao['POPULAÇÃO'].astype (int)
```

```python
populacao[populacao['POPULAÇÃO'].str.contains ('\(', na=False)]
```

```python
populacao = populacao.assign (populacao_tratado = populacao['POPULAÇÃO'].replace ('\(\d{1,2}\)', '', regex=True).replace ('\.', '', regex=True))
```

```python
populacao
```

```python
populacao[populacao['POPULAÇÃO'].str.contains ('\(', na=False)]
```

```python
populacao = populacao.astype ({'populacao_tratado':'int64'})
```

```python
populacao_estados = populacao.groupby ('UF')[['populacao_tratado']].sum ().reset_index()
populacao_estados
```

```python
emissao_estados = emissoes_por_ano[emissoes_por_ano['Ano'] == 2021].groupby ('Estado')[['Emissão']].sum().reset_index()
emissao_estados
```

```python
relacao_emissao_populacao = pd.merge (emissao_estados, populacao_estados, left_on='Estado', right_on='UF', how='inner')
```

```python
relacao_emissao_populacao
```

```python
relacao_emissao_populacao.plot.scatter (x='populacao_tratado', y='Emissão')
```

```python
import plotly.express as px
```

```python
px.scatter (data_frame=relacao_emissao_populacao, x='populacao_tratado', y='Emissão', text='Estado', opacity=0)
```

```python
relacao_emissao_populacao
```

```python
relacao_emissao_populacao = relacao_emissao_populacao.assign(emissao_per_capita = relacao_emissao_populacao['Emissão']/relacao_emissao_populacao['populacao_tratado']).sort_values('emissao_per_capita', ascending = False)
relacao_emissao_populacao
```
