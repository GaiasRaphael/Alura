```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
```

```python
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados ['Bairro']
```

```python
dados
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
def nota_extra (nota):
    return nota * 0.4


alunos['Ponto_extra'] = alunos['Notas'].apply(nota_extra)
alunos
```

```python
alunos['Notas_finais'] = alunos['Notas'] + alunos['Ponto_extra']
alunos
```

```python
alunos['Aprovado_final'] = alunos['Notas_finais'].apply(lambda x: "Aprovado"
                                                        if x >= 6.0 else
                                                        "Reprovado")
alunos
```
