```python
q_seguranca = 5
s_seguranca = 3000

q_docente = 16
s_docente = 6000

q_diretoria = 1
s_diretoria = 12500

dados = {
    s_seguranca: q_seguranca,
    s_docente: q_docente,
    s_diretoria: q_diretoria
}

```

```python
def mediaPonderada (dados):
    soma_ponderada = sum(valor * peso for valor, peso in dados.items())
    soma_pesos = sum(dados.values())
    return soma_ponderada / soma_pesos
```

```python
print(mediaPonderada(dados))
```
