```python
def media(lista: list=[0]) -> float:
  ''' Função para calcular a média de notas passadas por uma lista

  lista: list, default [0]
    Lista com as notas para calcular a média
  return = calculo: float
    Média calculada
  '''
  calculo = sum(lista) / len(lista)

  if len (lista) > 4:
    raise ValueError ("A lista não pode ter mais de 4 notas!")
  
  return calculo
```

```python
def lista_soma (lista1, lista2):
    if len (lista1) > len (lista2) or len (lista1) < len (lista2):
        raise ValueError("A quantidade de elementos em cada lista é diferente.")
    
    new_list = []
    for i in range (len(lista1)):
        tupla = (lista1[i], lista2[i], lista1[i] + lista2[i])
        new_list.append (tupla)

    return new_list
```

```python
def compara_lista (list1, list2):
    if len (list1) > len (list2) or len (list1) < len (list2):
        raise ValueError("A quantidade de elementos em cada lista é diferente.")
    
    resultado = 0
    for i in range (len (list1)):
        if ord (list1[i]) > 68:
            raise ValueError(f"A alternativa {list1[i]} não é uma opção de alternativa válida")
        
        if list1[i] == list2[i]:
            resultado += 1
        
    return resultado
        

def result_notas (lista):
    gabarito = ['D', 'A', 'B', 'C', 'A']
    
    for i in range (len(lista)):
        result = compara_lista (lista[i], gabarito)

        lista[i] = result
    
    return lista
    
```
